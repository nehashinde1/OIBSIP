```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly.express as px
dataset=pd.read_csv("Unemployment_Rate.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.columns=['States', 'Date', 'Frequency', 'Estimated Unemployment Rate',
       'Estimated Employed', 'Estimated Labour Participation Rate',
       'Region', 'longitude', 'latitude']
```

```python
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(dataset.corr(),cmap='BuPu')
plt.show()
```

```python
dataset.columns=['States', 'Date', 'Frequency', 'Estimated Unemployment Rate',
       'Estimated Employed', 'Estimated Labour Participation Rate',
       'Region', 'longitude', 'latitude']
plt.title("Unemployment Rate in India")
sns.histplot(x="Estimated Employed", hue="Region",data=dataset)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Unemployment Rate in India")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=dataset)
plt.show()
```

```python
unemploy=dataset[['States','Region','Estimated Unemployment Rate']]
fig=px.sunburst(unemploy,path=['Region','States'],
                values='Estimated Unemployment Rate',
                width=800,height=800,color_continuous_scale="brbg",
                title="Unemployment Rate in India")
fig.show()
```
